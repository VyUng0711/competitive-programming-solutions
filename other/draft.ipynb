```python
def check(board, row, col):
    # Check the rows:
    for i in range(row):
        if board[i][col]:
            return False
    # Check the two diagonals:
    i = row
    j = col
    while i >= 0 and j >= 0:
        if board[i][j]:
            return False
        i -= 1
        j -= 1
    i = row
    j = col
    while j < n and i >= 0:
        if board[i][j]:
            return False
        i -= 1
        j += 1
    return True
```

```python
def nqueen(board, row):
    if row == n:
        # printsolution()
        addsolution()
        # print()
        return True
    for j in range(n):
        if check(board, row, j) == True:
            board[row][j] = 1
            nqueen(board, row + 1)
            board[row][j] = 0
    return False
```
